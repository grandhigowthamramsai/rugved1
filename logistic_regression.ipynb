```python
Train,Valid,Test=np.split(df.sample(frac=1),[int(0.6*len(df)),int(0.8*len(df))])
```

```python
def mean(Train,x):
  y=Train[x].mean()

  y1=Train[x].std()
  for index, row in Train.iterrows():
    Train.at[index, x] = (row[x]-y)/y1


  return Train[x]
```

```python
df['Time_spent_Alone']=mean(df,'Time_spent_Alone')
df['Going_outside']=mean(df,'Going_outside')
df['Friends_circle_size']=mean(df,'Friends_circle_size')
df['Post_frequency']=mean(df,'Post_frequency')
```

```python
def seperate(df):
  x=df.iloc[:, :-1]
  y=df.iloc[:,-1]

  return x,y
train_x,train_y=seperate(Train)
valid_x,valid_y=seperate(Valid)
test_x,test_y=seperate(Test)
```

```python
class logistic_regression():
  def __init__(self,X,Y):
    self.X=X
    self.Y=Y
    self.sum_cols=0
    self.weights=None
    self.bias=None
  def weigh(self):
    self.weights_1D=np.zeros(8,)
    self.weights_2D = np.expand_dims(self.weights_1D, axis=1)

    bias=self.weights_2D[0]
  def sigmoidp(self,z):

    return 1/(1+math.exp(-z))



  def val(self):

    arr_X = self.X.to_numpy()
    arr_Y=self.Y.to_numpy()
    arr_Y = np.expand_dims(arr_Y, axis=1)

    X0_ARR=np.ones(self.X.shape[0],)
    X0_ARR_2D= np.expand_dims(X0_ARR, axis=1)
    arr_X=np.hstack((X0_ARR_2D,arr_X))


    epoch=0
    while(epoch<100):
      iteration=0

      while(iteration<len(arr_X)):
        for index in range(len(self.weights_2D)):

          FEATURE_2D= np.expand_dims(arr_X[iteration], axis=1)
          FEATURE_2D=np.transpose(FEATURE_2D)
          DOT_PROD=np.dot(FEATURE_2D,self.weights_2D)
          z=DOT_PROD[0][0]
          self.weights_2D[index][0]-=0.012*((self.sigmoidp(z)-arr_Y[iteration][0])*FEATURE_2D[0][index])

        iteration+=1

        if(iteration%100==0):

          print('iteration ',iteration )
          #print('BIAS  : ',self.weights_2D[0][0],'WEIGHTS 1  :  ',self.weights_2D[1][0],'WEIGHTS 2  :  ',self.weights_2D[2][0],'WEIGHTS 3  :  ',self.weights_2D[3][0])



      print(self.weights_2D)
      epoch+=1
  def sigmoid(self,z):
    x= 1/(1+math.exp(-z))
    if(x>=0.5):
      return 1

    else:
      return 0

  def test(self,X,Y):
    self.X=X
    self.Y=Y
    arr_X = self.X.to_numpy()
    arr_Y=self.Y.to_numpy()

    X0_ARR=np.ones(self.X.shape[0],)
    X0_ARR_2D= np.expand_dims(X0_ARR, axis=1)
    arr_X=np.hstack((X0_ARR_2D,arr_X))
    arr_Y= np.expand_dims(arr_Y, axis=1)
    iteration=0
    count=0

    self.pre=np.empty([arr_X.shape[0],1])
    print(self.pre.shape)
    print(self.weights_2D.shape)
    print(arr_X.shape)
    while(iteration<arr_X.shape[0]):
      z=np.dot(arr_X[iteration],self.weights_2D)

      self.pre[iteration][0]= self.sigmoid(z)

      if(self.pre[iteration][0]==arr_Y[iteration][0]):
        count+=1
      iteration+=1

    print(count/iteration)
  def plot(self,Y):
    arr_Y=self.Y.to_numpy()
    arr_Y = np.expand_dims(arr_Y, axis=1)
    print(arr_Y.shape,self.pre.shape )
    plt.scatter(arr_Y[:,0],self.pre[:,0] , color='red', label='Actual Values')
    plt.plot(arr_Y,arr_Y, color='blue', label='Predicted Line')
    plt.show()
```
